and, or, not are boolean operators to compare boolean values

and operator takes two boolean values (or expressions) so its considered a boolean operator. and operator evalutes an expression to true if both boolean values are true: otherwise it evalutes to false

```python
True and True
```

```python
True and False
```

or operator evalutes an expression to true if either of the boolean values is true

```python
False or True
```

```python
False or False
```
