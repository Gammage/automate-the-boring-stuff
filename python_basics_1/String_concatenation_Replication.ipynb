Operators change base on data types. see examples

+ used to combine 2 string values it joins as the string concatenation

```python
'alice' + 'bob'
```

+ used in string and integer will cause typeError

```python
'alice' + 42
```

* multiples two integer or floating-point values. * used on one str and one int, it becomes string replication operator

```python
'scott' * 5
```

can only be used with two numeric values or one str and one int value (for sting replication)
